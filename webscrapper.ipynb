```python
import pandas as pd
import os
from tkinter import image_names
from GoogleImageScraper import GoogleImageScraper
```

```python
%load_ext autoreload
%autoreload 2
```

```python

webdriver_path = os.path.normpath(os.path.join(os.getcwd(), 'webdriver\chromedriver.exe'))
image_path = os.path.normpath(os.path.join(os.getcwd(), 'photos'))
search_keys= ["kaya toast","roti prata","ice kacang","pandan cake", "Nasi lemak"]
number_of_images = 500
data=pd.DataFrame(columns=["label","image_url","filename"])
for search_key in search_keys:
    image_scraper = GoogleImageScraper(webdriver_path,image_path,search_key,number_of_images)
    image_urls = image_scraper.find_image_urls()
    a=image_scraper.save_images(image_urls, keep_filenames=False)
    for i in range(len(a)):
        data2=pd.DataFrame({
            'label':[search_key],
            'image_url':[image_urls[i]], 
            'filename':[a[i]]
        })
        data=data.append(data2,ignore_index=True)
data.to_csv('original.csv')
```
